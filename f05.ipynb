```python
with open(wrong_golden, "r", encoding="utf-8") as w_g:
    with open(wrong_predicted, "r", encoding="utf-8") as w_p:
        w_g_content = w_g.readlines()
        w_p_content = w_p.readlines()

        w_g_filtered = {}
        i = 0

        while i < len(w_g_content):
            phrase = w_g_content[i]
            i+= 1
            current_phrase = []
            while w_g_content[i] != '\n':
                if w_g_content[i] != 'A -1 -1|||noop|||-NONE-|||REQUIRED|||-NONE-|||0\n':
                    current_phrase.append(w_g_content[i])
                #else:
                    #print("found one?")
                i += 1
            w_g_filtered[phrase] = current_phrase
            i+=1

        w_p_filtered = {}
        i = 0

        while i < len(w_p_content):
            phrase = w_p_content[i]
            i+= 1
            current_phrase = []
            while w_p_content[i] != '\n':
                if w_p_content[i] != 'A -1 -1|||noop|||-NONE-|||REQUIRED|||-NONE-|||0\n':
                    current_phrase.append(w_p_content[i])
                #else:
                #    print("found one?")
                i += 1
            w_p_filtered[phrase] = current_phrase
            i+=1


        TP=0
        FP=0
        FN=0
        for key in w_p_filtered:
            for item_predict in w_p_filtered[key]:
                #if item_predict != 'A -1 -1|||noop|||-NONE-|||REQUIRED|||-NONE-|||0\n':
                    if item_predict in w_g_filtered[key]:
                        #print(item_predict)
                        TP+=1
                    else:
                        FP+=1

            for item_golden in w_g_filtered[key]:
                #if item_golden != 'A -1 -1|||noop|||-NONE-|||REQUIRED|||-NONE-|||0\n':
                    if item_golden not in w_p_filtered[key]:
                        FN+=1

        P = TP/(TP + FP)
        R = TP/(TP + FN)

        F05 = (1.25 * TP)/(1.25*TP + 0.25*FN + FP) #1.25* ((P * R)/((0.25*P) + R))


        print(P)
        print(R)
        print(F05)
```
